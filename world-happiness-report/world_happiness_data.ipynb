```python
import pandas as pd
import matplotlib.pyplot as plt
from pprint import pprint

happiness2015 = pd.read_csv("2015.csv")
first_5 = happiness2015[:5]
happiness2015.info()
```

```python
mean_happiness = {}
regions = happiness2015['Region'].unique()

for region_group in regions:
    region_mean = happiness2015[happiness2015['Region']==region_group]['Happiness Score'].mean()
    mean_happiness[region_group] = region_mean
pprint(mean_happiness)
```

```python
so_asia = happiness2015[happiness2015['Region'] == 'Southern Asia']
so_asia.plot(x='Country', y='Happiness Score', kind='barh', title='Southern Asia Happiness Scores', xlim=(0,10))
```

```python
plt.barh(list(mean_happiness.keys()), mean_happiness.values())
plt.xlim
plt.show()
```
