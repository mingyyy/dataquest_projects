```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
```

```python
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255,128/255,14/255)
cb_lightgrey = (171/255, 171/255, 171/255)
```

```python
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
plt.plot(women_degrees['Year'], women_degrees['Biology'], label='Female%', c='r')
plt.legend()
plt.show()
```

```python
plt.plot(women_degrees['Year'], women_degrees['Biology'], c='blue', label='Women')
plt.plot(women_degrees['Year'], 100-women_degrees['Biology'], c='green', label='Men')
plt.legend(loc='upper right')
plt.title('Percentage of Biology Degrees Awarded By Gender')
plt.tick_params(bottom=False, top=False, left=False, right=False)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(women_degrees['Year'], women_degrees['Biology'], label='Women')
ax.plot(women_degrees['Year'], 100-women_degrees['Biology'], label='Men')

ax.tick_params(bottom=False, top=False, left=False, right=False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.set_title('Percentage of Biology Degrees Awarded By Gender')
ax.legend(loc="upper right")

plt.show()
```

```python
major_cats = ['Biology', 'Computer Science', 'Engineering', 'Math and Statistics']
fig = plt.figure(figsize=(12, 12))

for sp in range(0,4):
    ax = fig.add_subplot(2,2,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[major_cats[sp]], c=cb_orange, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[major_cats[sp]], c=cb_dark_blue, label='Men', linewidth=2)
    # Add your code here.
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    for spine in ax.spines.values():
        spine.set_visible(False)
    ax.set_title(major_cats[sp])

# Calling pyplot.legend() here will add the legend to the last subplot that was created.
plt.legend(loc='upper right')
plt.show()
```

```python
fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange,linewidth=3)
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom=False, top=False, left=False, right=False)

plt.show()
```

```python
fig = plt.figure(figsize=(10, 15))

for col in range(0,3):
    for row in range(0,6):
        ax = fig.add_subplot(6,3,(col+1)+3*(row))
        
        ax.spines["right"].set_visible(False)    
        ax.spines["left"].set_visible(False)
        ax.spines["top"].set_visible(False)    
        ax.spines["bottom"].set_visible(False)
        ax.set_xlim(1970, 2010)
        ax.set_ylim(0,100)
        ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
        
        if col != 1 and row !=5:
            ax.set_yticks([0, 50, 100])
            ln=ax.axhline(50, c=cb_lightgrey, alpha=0.3)
        
        if col == 0:
            ax.plot(women_degrees['Year'], women_degrees[stem_cats[row]], c=cb_orange, label='Women', linewidth=2)
            ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[row]], c=cb_dark_blue, label='Men', linewidth=2)
            ax.set_title(stem_cats[row])
            if row == 0:
                ax.text(2003, 87, 'Women')
                ax.text(2005, 8, 'Men')
            elif row == 5:
                ax.text(2003, 8, 'Women')
                ax.text(2005, 87, 'Men')
                ax.tick_params(labelbottom=True)
        if col == 1 and row < 5:
            ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[row]], c=cb_orange, label='Women', linewidth=2)
            ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[row]], c=cb_dark_blue, label='Men', linewidth=2)
            ax.set_title(lib_arts_cats[row])
            if row == 0:
                ax.text(2003, 80, 'Women')
                ax.text(2005, 18, 'Men')
            elif row == 4:
                ax.tick_params(labelbottom=True)
           
        if col == 2:
            ax.plot(women_degrees['Year'], women_degrees[other_cats[row]], c=cb_orange, label='Women', linewidth=2)
            ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[row]], c=cb_dark_blue, label='Men', linewidth=2)
            ax.set_title(other_cats[row])
            if row == 0:
                ax.text(2003, 92, 'Women')
                ax.text(2005, 2, 'Men')
            elif row == 5:
                ax.text(2003, 62, 'Women')
                ax.text(2005, 35, 'Men')
                ax.tick_params(labelbottom=True)
            
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.25, hspace=0.25)
plt.savefig('biology_degrees.png')
plt.show()
```
